# Variables


Variables, constants and keywords are some of the basic building blocks in the python language.   There are specific requirements with regards to variable syntax or naming conventions when creating a variable.  A constant, which is a fixed value variable is also important in that the value does not change and will give you an error if you try to modify it. A requirement for a constant is that the name be in upper case.

Variables and constants can have both letters and numbers but should not start with numbers ,contain an illegal character such as &,! or @ or be a python keyword such as the following:

and, except, lambda, with, as, finally, nonlocal, while, assert, false, None, yield, break, for, not, class, from, or, continue, global, pass, def, if, raise, del, import, return, elif, in, True, else, is, try

Variable data types include integer, float and string.  One other consideration when creating a variable would be to use a meaningful name that would be applicable towards what the variable will be used for and the value it will contain.

```python
name = "Christie"
type(name)
```
