##Functions

Functions are a useful tool in python programming as with any other programming language.  Funtions allow the following:

!. Improved clarity since the code is built in a logical sequence of statements.

2. Improved usability since instead of having code having code replicated anyplace it is required in a program, you can just have the function code defined in one place within a program and then call the function from anywhere its use is needed.
3.  From a program user perspective, functions allow for abstraction or for the user to call a defined function knowing what arguments its takes and what the result is without having to know how the computation within the function is done.

If you are not familiar with any of the concepts outline above, dont worry, each will be explained.


Functions in python include:
1. User defined function
2. Build in function.


##User defined function

Functions in python having the following structure:
 
def function_name ():  
  ----statement(s) 

The first line in the function is the function definition which includes the keyword def, the function_name, parathesis - which may or may not include parameters, ending with the colon :.  After the function definition is the function body where will exist one or more statments that are each indented four spaces and perform any computations that are outlined.  

A function that does not use any parameters is called a void function since only that which is in the function body gets executed.  In the case a print statment is executed with nothing returned to the calling statement.

```python
def print_name ():  #function definition
  print("John Doe") #function body

print_name() #statement calling function. 
```

An example of a function that does takes an argument but does not return a result to the caller.

```python
def print_full_name (name): #name variable used as the arguement within the function definition
  print("Your name is " + name)

print_full_name("John Doe") # String with individual name passed as a parameter within the function caller
```

Now an example of a function that takes an argument, utilizes the argument within some typy of computaton in the function body then returns a result to the function caller.  In this example will convert the temperature celcius to fahrenheit.

```python
def celc_to_fahrenheit_converter (celcius):
   result = (celcius * 1.8) + 32
   return result

fahrenheit = celc_to_fahrenheit_converter(36) #call to function with the argument value of 36

print(fahrenheit)
```

Another example would be a function calling a function.  The previous function print_full_name will be called multiple times.

```python
def multi_print (c):
  print_full_name(c)
  print_full_name(c)
  print_full_name(c)
  print_full_name(c)

multi_print("John Doe")

```

##Built In Functions

Lets say you instead of just printing the value of fahrenheit, 98.6, you would like to print out a string such as 98.6 degrees Fahrenheit. The structure would be as follows:

print(str(variable) + " string"). 

One thing to notice is the addition of the built in function str() or string method which converts any number to a string.  The use of the plus operator concantenates the variable with the string in double quotes.  A last note to make here is a space is used between the first double quote and the first character of the string.  This is required so there will be a space between the value and the rest of the string so it becomes a normal sentence.

This is the code showing this structure:

```python
print(str(fahrenheit) + " degrees fahreneheit")
```

If you had not used the string method str() to convert the number to a string, the following error would occur as can be seen when the following code is run:

print(fahrenheit + " degrees fahrenheit)

As seen in the error statement above, the TypeError details that the operand + can not be used to concatenate a float and str variable.  This is why the call is made to the built string method using the variable fahrenheit, with its float value 98.6.  The string value "98.6" is returned.

Other built in functions that you have already seen in use are range() and print().  One other useful function is input(), which takes user input from the keyboard then does some computation. Lets try the following code example where the use is asked to enter temperature in Celcius in order to convert to and print in Fahrenheit:

```python
user_input = input("Enter you temperature in Celcius ")
result = (user_input * 1.8) + 32
print(str(result))
```

When the code block is run the following error occurs:

Input Function Error Screenshot.png

As can be seen above, the arrow ----> is pointing to the second line of code.  In particular the TypeError is stating it can't multiply the sequence by the type float.  Here sequence reference to the value entered by the user when prompted by the program.  In this case the value entered is considered a string also called a sequence since it is a set of characters.  In order to convert the entered string or sequence into a numeric value that be part of the calculation user_input * 1.8, one might want to use the built in function int() or the integer method.  In the event though that a value of 36.6 might be entered instead of 36 it would be appropriate to use the float method float() so that all such numberical values can be converted.  

The restructuring of the statement would then be result = (float(user_input) * 18) where float uses user_input as its argument in the function call.

Ok, now that the rationale for this conversion has been covered let's rerun the block of code.

```python
user_input = input("Enter you temperature in Celcius ")
result = ((float(user_input) * 1.8) + 32) 
print(str(result))
```

Depending on the value you entered whether it was something like 36 or 38.5, you may get 2 or more numbers after the decimal place such as 100.43 or 97.230000000.  In order to round the result to once decimal place the function round() is required.  The function takes two arguments, the value or variable holding the value in addition to the number of decimal places to round to such as round(variable, number).  Let's modify the previous code block to demonstrate this:

```python
user_input = input("Enter you temperature in Celcius ")
result = round(((float(user_input) * 1.8) + 32), 1)
print(str(result))
```

One other concept that should be covered here is error handling using try /except.  Error handling covers exceptions that occur during the execution of a code block within a program.  One type of exception that has already be covered ia a TypeError.  Other types include ZeroDivisionError and ValueError.

In the event that you had entered non numeric value inadvertently at the prompt for the temperature in Celcius such at the string "thirty seven" the code block will throw the following ValueError exception:

Value Error Exception Screenshot.png

When reimplementing the code block with a try/except structure, a while block will be added in addition to a break statement at the end of the try block.  The break statment will allow the while loop to exit once a numeric value is entered. If the break statment had not been added to the try block, once a numeric value is entered the loop would run infinitely.

```python
user_input = input("Enter you temperature in Celcius: ")
while True:
  try:
    result = round(((float(user_input) * 1.8) + 32), 1)
    print(str(result))
    break
    
  except:
    print("You did not enter a numeric value.")
    user_input = input("Enter you temperature in Celcius: ")
```
